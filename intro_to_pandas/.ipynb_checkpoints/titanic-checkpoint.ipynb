```python
sex_survived = titanic[["Sex","Survived"]]
```

```python
sex_survived.head(8)
```

```python
# to find total male and female survived 
survived_female = sex_survived[(sex_survived["Sex"] == "female")]["Survived"].sum()
total_female = (titanic["Sex"] == "female").sum()
survived_male = sex_survived[(sex_survived["Sex"] == "male")]["Survived"].sum()
total_male = (titanic["Sex"] == "male").sum()
print("Out of ", total_male ," males ", survived_male, " survived") 
print("Out of ", total_female ," female ", survived_female, " survived") 
```

## But loc method is very helpful to select specific row and column



```python
survived_female = titanic.loc[titanic["Sex"] == "female", "Name"]
```

```python
survived_female.count()
```

```python
titanic.iloc[9:25, 2:5]
```


# REMEMBER

    When selecting subsets of data, square brackets [] are used.

    Inside these brackets, you can use a single column/row label, a list of column/row labels, a slice of labels, a conditional expression or a colon.

    Select specific rows and/or columns using loc when using the row and column names.

    Select specific rows and/or columns using iloc when using the positions in the table.

    You can assign new values to a selection based on loc/iloc.
